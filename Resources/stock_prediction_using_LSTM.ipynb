# 4. Data Preprocessing

```python
df.shape
```

```python
data_training = pd.DataFrame(df.Close[0:int(len(df)*0.7)])
data_testing = pd.DataFrame(df.Close[int(len(df)*0.7): int(len(df))])

print(data_training)
print(data_testing)
```

```python
data_testing
```

# Scaling down the data between 0 & 1

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array = scaler.fit_transform(data_training)
data_training_array
```

```python
type(data_training_array)
```

```python
data_training_array.shape
```

# 5. Sequence Creation

```python
12, 13, 12, 15, 10, 14, 16, 18, 19, 20 
```

```python
x_train = []
y_train = []

for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100: i])
    y_train.append(data_training_array[i,0])

x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train
```

```python
y_train
```

```python
x_train.ndim
```

```python
y_train.ndim
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Input
```

```python
model = Sequential()

model.add(Input(shape=(100, 1))) #100 data taken taken, 1 Close
model.add(LSTM(units=128, activation='tanh', return_sequences=True)) # 128 nuraons for LSTM, default hyperbolic tangent, have return sequence 
model.add(LSTM(units=64)) #summary of above return sequnce
model.add(Dense(25)) # Analiyse it in 25 different angle
model.add(Dense(1)) # Final One output
```

# 7. Model Taining

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, epochs=50)
```

```python
model.summary()
```

```python
model.save('sock_prediction_model.keras')
#save the trained model
```

# 10. Model Evaluation

```python
 # Mean Squared Error (MSE)
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_predicted)
print(f"Mean Squared Error (MSE): {mse}")
```

```python
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE) : {rmse}")
```

```python
r2 = r2_score(y_test, y_predicted)
print(f"R-Squared: {r2}")
```
