## Bar Chart

```python
months = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [100, 200, 300, 400]
```

```python
plt.figure(figsize=(10,6))
plt.bar(months, revenue)
plt.title('Monthly Revenues')
plt.xlabel('Months')
plt.ylabel('Revenue')
```

## Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
orignal_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 24, 28, 37]
```

```python
plt.plot(days, orignal_price, 'b', label='Original Price')
plt.plot(days, predicted_price, 'r', label='Predicted Price')
plt.legend()
plt.title("Stock Price")
plt.xlabel("Weekdays")
plt.ylabel('Price')
```

```python
plt.subplot(3, 1, 1)
plt.plot(days, orignal_price, 'b-o', label='Orignal Price')
plt.title('Orignal Stock Price')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(days, predicted_price, 'r-o', label='Predicted Price')
plt.title('Predicted Stock Price')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

price_differece = [ o-p for o, p in zip(orignal_price, predicted_price)]
plt.subplot(3, 1, 3)
plt.bar(days, price_differece, color='purple' , label='Predicted Price')
plt.title('Differences Predicted Stock Price')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend()

plt.tight_layout()
```

```python
plt.plot(days, orignal_price, 'b', label='Original Price')
plt.plot(days, predicted_price, 'r', label='Predicted Price')
plt.legend()
plt.title("Stock Price")
plt.xlabel("Weekdays")
plt.ylabel('Price')
plt.savefig('example_prediction.png', dpi=300)
```
